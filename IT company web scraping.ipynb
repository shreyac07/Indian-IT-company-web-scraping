```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/Big_Tech_(India)'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find('table')
```

```python
table = soup.find_all('table')[0]
```

```python
india_titles = table.find_all('th')
```

```python
india_titles
```

```python
india_table_titles = [title.text.strip() for title in india_titles]
print(india_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = india_table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = individual_row_data
```
